## Let's join the tables in satisfaction of the Project 3 question prompt
Customers’ Last Name and First Name, and each customer’s purchased Track names and Album Titles. The information should be sorted by Customer LastName then Customer FirstName. 

### The first step is to select only the relevant columns that are necessary for us to join the the dataframes

```python
df_customer = customer[['CustomerId', 'LastName', 'FirstName']]
df_customer.head()
```

```python
df_invoice = invoice[['CustomerId', 'InvoiceId']]
df_invoice.head()
```

```python
df_invoice_line = invoice_line[['InvoiceId', 'TrackId']]
df_invoice_line.head()
```

```python
df_track = track[['TrackId', 'Name', 'AlbumId']]
df_track.head()
```

```python
df_album = album[['AlbumId', 'Title']]
df_album.head()
```

### The second step is to begin to merge the tables.

```python
# Merge output [19] and output[20] on CustomerId
temp1 = pd.merge(df_customer, df_invoice, how='inner', on='CustomerId')
temp1.head()
```

```python
# Merge output [24] and output [21] on invoiceId as both tables have the column in common
temp2 = pd.merge(temp1, df_invoice_line, how='inner', on='InvoiceId')
temp2.head()
```

```python
# Merge output [25] and output [22] on Track Id as they both have the columns in common.
temp3 = pd.merge(temp2, df_track, how='inner', on='TrackId')
temp3.head()
```

```python
# Merge output [26] and output [23]
final = pd.merge(temp3, df_album, how='left', on='AlbumId')
final.head()
```

### Lastly, remove the columns that are no longer needed.

```python
df_final = final[['LastName', 'FirstName', 'Name', 'Title']].sort_values(by=['LastName', 'FirstName'])
df_final.head(10)
```

```python
df_final = df_final.reset_index(drop=True)
```
